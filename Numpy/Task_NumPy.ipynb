```python
import numpy as np
```

```python
a = np.array([[1, 2, 1],
              [-2, 1, 1]])

b = np.array([[3, 1],
              [-1, 0],
              [2, 4]])

c = np.array([[-2, 2],
              [1, -1],
              [1, 2]])
```

#TASK 1:
*Выполнить операции над матрицами: <br>
1) A+2*B <br>
*2) B+2*C

```python
'''A+2*B'''
b2 = np.dot(2, b)
try:
    print(a + b2)
except ValueError:
    print('A+2*B:\nFold the Matrix of different sizes is prohibited')
```

```python
'''B+2*C'''
c2 = np.dot(2, c)
try:
    print('B+2C:\n', b + c2)
except ValueError:
    print('B+2*C: fold the Matrix of different sizes is prohibited')
```
