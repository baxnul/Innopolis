```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Titanic-Dataset.csv', index_col='PassengerId') # Прочитали файл csv
# df = df.set_index('PassengerId') # Переопределили строку индексации данных на свою
df
```

```python
df.info()
```

```python
df.isnull().sum().sum() # Количество пустых ячеек во всей таблице
```

```python
df['Age'] = df['Age'].fillna(df.mean(axis=1)) # Замена пустоты в колонке Age средним значением
df
```

```python
df = df.fillna(method='ffill') # Для остальных колонок заполнили пустоты предыдущим значением
df.head()
```

```python
df = df.fillna(method='bfill') # Для остальных колонок заполнили пустоты следующим значением
df1 = df
df.info()
```

```python
uniques = pd.unique(df.values.ravel()) # Уникальные значения всей таблицы
len(uniques) # Кол-во уникальных значений
```

```python
df.describe() # Статистика по всем столбцам кол-во строк,  среднее значение, стандартные отклонения значений в кадре данных, минимальное
```

```python
# Получим максимальные значения в каждом столбце
maximums = df.max()
maximums
```

```python
df['Age'].max() #Максимум в столбце Age
```

```python
# Получим минимальные значения в каждом столбце
minimums = df.min()
minimums
```

```python
df['Age'].min() # Минимум в столбце Age
```

```python
# Конвертируем колонку с возрастом в тип Int
df1['Age'] = df1['Age'].astype(int)
df1['Age'].head()
```

```python
# Разбиваем возраст на интервалы в новой колонке по 10 лет(до 90 лет)
df1['Age Group'] = pd.cut(df1['Age'], bins=range(0, 91, 10))
# df1['Age Group']
df1.head()
```

```python
df1['Survived'].value_counts(dropna=False)  #Кол-во погибших/выживших, dropna=False это не сбрасывать незаполненные значения
```

```python
# # Выживаемость по Годам (Age)
survived_died_counts = df1.groupby(['Age', 'Survived']).size().unstack()
survived_died_counts
```

```python
# Группировка выживших/не выживших по возрастам, с интервалом в 10 лет
survived_died_counts = df1.groupby(['Age Group', 'Survived']).size().unstack()
survived_died_counts['sum_people'] = survived_died_counts.sum(axis=1) # всего человек в интервалах в новой колонке sum_people
survived_died_counts
```

*Вывод: больше всего выжило в процентном соотношении людей в возрасте 30-40 лет*

```python
# Выживаемость по Классу Круиза (Pclass)
survival_Pclass = df.groupby('Pclass')['Survived'].value_counts().unstack(fill_value=0)
survival_Pclass['sum_people'] = survival_Pclass.sum(axis=1)
survival_Pclass
```

*Вывод: больше всего выжило людей в 1 классе*

```python
# # Выживаемость по половому признаку (sex)
survival_by_sex = df.groupby('Sex')['Survived'].value_counts().unstack(fill_value=0)
survival_by_sex['total_people'] = survival_by_sex.sum(axis=1)
survival_by_sex
```

*Вывод: больше всего выжило женского пола*

```python
# # Выживаемость по числу братьев,сестер или супругов на борту у человека (SibSp)
survival_by_SibSP = df.groupby('SibSp')['Survived'].value_counts().unstack(fill_value=0)
survival_by_SibSP['sum_people'] = survival_by_SibSP.sum(axis=1)
survival_by_SibSP
```

*Вывод: чем меньше братьев тем больше шанс выжить*

```python
# Выживаемость по количеству родителей или детей, с которыми путешествовал каждый пассажир (Parch)
survival_by_Parch = df.groupby('Parch')['Survived'].value_counts().unstack(fill_value=0)
survival_by_Parch['Count_people'] = survival_by_Parch.sum(axis=1)
survival_by_Parch
```

*Вывод: больше всего выжило людей кто был без родителей и детей*

```python
# Выживаемость по номеру билета (Ticket)
survival_by_Ticket = df.groupby('Ticket')['Survived'].value_counts().unstack(fill_value=0)
survival_by_Ticket['count_people_buy'] = survival_by_Ticket.sum(axis=1)
survival_by_Ticket
```

```python
# Выживаемость по цене билета (Fare)
survival_by_Fare = df.groupby('Fare')['Survived'].value_counts().unstack(fill_value=0)
survival_by_Fare['Count_people_buy'] = survival_by_Fare.sum(axis=1)
survival_by_Fare
```

```python
# Выживаемость по номеру каюты (Cabin)
survival_by_Cabin = df.groupby('Cabin')['Survived'].value_counts().unstack(fill_value=0)
survival_by_Cabin['sum_people'] = survival_by_Cabin.sum(axis=1)
survival_by_Cabin
```

```python
# Выживаемость по порту посадки (Embarked)
survival_by_Embarked = df.groupby('Embarked')['Survived'].value_counts().unstack(fill_value=0)
survival_by_Embarked['sum_people'] = survival_by_Embarked.sum(axis=1)
survival_by_Embarked
```

*Вывод: больше всего выжило людей кто был из порта С (в % соотношении)*
